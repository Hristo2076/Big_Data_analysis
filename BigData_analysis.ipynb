#**Python_pandas**

```python
import pandas as pd
```

 ## **Download DataFrame**

```python
columns = ['user_id', 'item_id', 'rating', 'date']
ratings = pd.read_csv('u.data.csv', sep='\s+', header = None, names = columns)
ratings.head()
```

```python
columns = ['movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy',
           'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film_Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci_Fi',
           'Thriller','War','Western']
films = pd.read_csv('item.csv', sep='|', encoding='latin-1', header = None, names = columns).drop('video_release_date',axis=1)

films.head()
```

**The most active user**

```python
ratings[['user_id','item_id']].groupby('user_id').count().sort_values('item_id', ascending=False).head(1)
```

actions of this user

```python
user_405 = ratings.loc[(ratings.user_id == 405)]
user_405[:737:73]
```

```python
film_user_405 = user_405.merge(films,left_on = 'item_id',right_on = 'movie_id')
film_user_405[:50:5]
```

**total number of rating**

```python
count_of_rating = ratings[['item_id','rating']].groupby('item_id').count()
count_of_rating.head(10)
```

**total sum of rating**

```python
sum_of_rating = ratings[['item_id','rating']].groupby('item_id').sum()
sum_of_rating.head(10)
```

```python
count_sum_item = count_of_rating.merge(sum_of_rating,on='item_id')
count_sum_item= count_sum_item.rename(columns={'item_id':'item_id',
                               'rating_x':'count_rating',
                               'rating_y':'sum_rating'})
```

```python
count_sum_item[:5]
```

```python
film_user_405 = film_user_405.merge(count_sum_item, on='item_id')
```

**join all necessary coumns**

```python
film_user_405 = film_user_405.drop(['movie_title','IMDb_URL','release_date'],axis=1)
film_user_405[:50:5]
```

#**Bilding a Linear_model**

**train,test, split moments**

```python
from sklearn.model_selection import train_test_split

X = film_user_405.drop('rating',axis=1)
y = film_user_405['rating']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,
                                                 train_size=0.67,
                                                 test_size=0.33,
                                                 random_state=42)
```

**Learn,learn more**

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.metrics import mean_squared_error
```

```python
clf = RidgeClassifier().fit(X_train,y_train)
clf
```

```python
y_pred = clf.predict(X_test)
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
import findspark
findspark.init()
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").getOrCreate()
```

```python
from pyspark.sql import functions as f
```

```python
data = spark.read.csv("u.data.csv", inferSchema= True, header= False, sep = '\t' ).withColumnRenamed('_c0','user_id')\
                                                                                  .withColumnRenamed('_c1','movie_id')\
                                                                                  .withColumnRenamed('_c2','rating')\
                                                                                  .withColumnRenamed('_c3','date')
```

```python
data.show(5)
```

```python
item = spark.read.csv("item.csv", inferSchema= True, header= False, sep = '|').withColumnRenamed('_c0','movie_id_item').withColumnRenamed('_c1','title').withColumnRenamed('_c2','Release_date')\
                                                                              .withColumnRenamed('_c3','video_release_date').withColumnRenamed('_c4','IMDb_URL').withColumnRenamed('_c5','unknown')\
                                                                              .withColumnRenamed('_c6','Action').withColumnRenamed('_c7','Adventure').withColumnRenamed('_c8','Animation')\
                                                                              .withColumnRenamed('_c9','Children').withColumnRenamed('_c10','Comedy').withColumnRenamed('_c11','Crime')\
                                                                              .withColumnRenamed('_c12','Documentary').withColumnRenamed('_c13','Drama').withColumnRenamed('_c14','Fantasy')\
                                                                              .withColumnRenamed('_c15','Film_Noir').withColumnRenamed('_c16','Horror').withColumnRenamed('_c17','Musical')\
                                                                              .withColumnRenamed('_c18','Mystery').withColumnRenamed('_c19','Romance').withColumnRenamed('_c20','Sci_Fi')\
                                                                              .withColumnRenamed('_c21','Thriller').withColumnRenamed('_c22','War').withColumnRenamed('_c23','Western')
```

```python
item.show(5)
```

**show on average score for film**

```python
data_item = data.join(item,data.movie_id == item.movie_id_item,'inner').drop('movie_id_item')
```

```python
data_item.show(5)
```

```python
data_item.groupBy('movie_id','title').agg(f.avg('rating').alias('avg')).show(5)
```

**show on average score for genre**

```python
item_genre = item.selectExpr('movie_id_item','title', "stack(19, 'unknown', unknown,'Action', Action,'Adventure',\
                            Adventure,'Animation',Animation, 'Children', Children, 'Comedy',\
                            Comedy, 'Crime', Crime, 'Documentary', Documentary,'Drama', Drama, \
                            'Fantasy', Fantasy, 'Film_Noir', Film_Noir, 'Horror', Horror, 'Musical', Musical, 'Mystery', Mystery, 'Romance',\
                            Romance, 'Sc_iFi', Sci_Fi, 'Thriller', Thriller, 'War', War, 'Western', Western) as (Genre, Flag)").where("Flag !=0")
```

```python
item_genre.show(10)
```

```python
data_item_genre = item_genre.join(data,data.movie_id == item_genre.movie_id_item, 'left').drop('movie_id')
```

```python
data_item_genre.show(10)
```

```python
data_item_genre.groupBy('Genre').agg(f.avg('rating').alias('avg_rating_genre')).show()
```

```python
top_5_popular = rating_count.orderBy('count_rating',ascending=False).show(5)
```
